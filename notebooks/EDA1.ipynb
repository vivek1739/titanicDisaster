#-----Summary Statistics------------
##1. Centrality measure : one number to represent entire set of values
    a) mean/avg - (total of n)/ n
 problem with mean : affected by extreme values
    b) median- middle value in sorted list
   ie. half values below this value and half above this value
   -> if list is even .'. take average of 2 middle values
#### NOTE : For Numerical value Centrality Measure(mean/median) and Dispersion measure or spread(range,percentile,variance,stand dev) . For categorical features ( total count, unique count, count per category, also we can group categorical data with numberical data)

##2. Spread/Dispersion Measure : how spread out values are from the central value
    ie. how similar and dissimilar data is there from central value
    Note : if data is similar, all will be around central value and if very dissimilar then it will be spread out from central value
-> Commonly used dispersion measure
    a) Range: Difference between maximum and minimum ie. range low= values packed together
            range high= values are spread out
     problem : affected by extreme values
    b) Percentiles: ex. x percentile is y means x% of values are below y
    ex. 50 percentile is 10 means 50% of values below 10
    -> percentile can help make quarters eg. 0-25 ,25-50 etc as buckets called QUARTILES
    
    NOTE: BOX-WHISKER PLOT : display the percentiles
  -> make 4 quarters min-25,25-50,50-75,75-max ( min is not zero as we exclude extreme values)
  -. make a box with 25-75 and make whiskers min-25 and 75-max 
  -. OUTLIERS are dots beyond min and max and are beyond 1.5 times IQR
  -. 50th percentile denotes median
  -. it can be made using pandas
  NOTE: Inter-quartile range ( diff b/w 75th and 25th percentile)
  
  c) Variance : how far each value is from mean value
      - small variance = less spread
        high variance = large spread
      variance = sum((value -mean)^2) / count

    -. unit is not clear
   .'. Standard Deviation = sq root ( variance)
   -. so unit same as that of the feature
   

# Summary Statistics

```python
# get statistics for all numerical columns
# gives many summary statistics
df.describe()
```

```python
# summary statistics one by one
## 1. Numerical Features
### a. centrality measure
# NOTE :
print('Mean fare : {0}'.format(df.Fare.mean()))
print('Mean fare : {0}'.format(df.Fare.median()))
```

```python
# dispersion measures
print('Min Fare : {0}'.format(df.Fare.min()))
print('Max Fare : {0}'.format(df.Fare.max()))
print('Fare Range : {0}'.format(df.Fare.max() - df.Fare.min()))
print('25 percentile : {0}'.format(df.Fare.quantile(.25)))
print('50 percentile : {0}'.format(df.Fare.quantile(.50)))
print('75 percentile : {0}'.format(df.Fare.quantile(.75)))
print('Variance fare : {0}'.format(df.Fare.var()))
print('Standard deviation fare : {0}'.format(df.Fare.std()))
```
